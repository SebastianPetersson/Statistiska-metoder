```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.optimize import minimize


plt.style.use('dark_background')
rng = np.random.default_rng(seed=42)
def histogram(data, ax, bins= None):
    ax.hist(data, bins=bins, edgecolor = 'black')  
    ax.set_title('Histogram')

def ogive(data, ax, bins=None):
    ax.hist(data, cumulative=True, bins=bins, color='orange', edgecolor='black') 
    ax.set_title('Ogive')
  
```

```python
X = rng.choice([1,0], size=1000, p=[0.4, 0.6])
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
X = rng.geometric(size=1000, p=0.3)
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
X = rng.binomial(n=100, p=0.5, size=1000)
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
X = rng.negative_binomial(n=100, p=0.5, size=1000)
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
X = rng.normal(size=1000, loc=50)
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
X = rng.standard_normal(size=1000)
fig, ax = plt.subplots(1, 2, figsize=[12,4])
histogram(X, ax[0])
ogive(X, ax[1])
```

```python
x = df['TV'].values
y = df['Sales'].values

a, b = np.polyfit(x, y, 1)
y_pred = a*x + b
residuals = y - y_pred
n = len(x)
s_err = np.sqrt(np.sum(residuals**2) / (n-2))
x_mean = np.mean(x)
se_line = s_err * np.sqrt(1/n + (x - x_mean)**2 / np.sum((x - x_mean)**2))
t = st.t.ppf(0.975, df=n-2)

lower = y_pred - t * se_line
upper = y_pred + t * se_line


plt.figure(figsize=(16,6))
plt.scatter(df['TV'], df['Sales'], s=5)
plt.plot(x, a*x+b, color = 'white', linestyle = ':')
plt.fill_between(x, lower, upper, color = 'red')
```

Rknar nedan ut mean  och st dev av Sales-kolumnen manuellt, fr att pvisa att df['Sales'].describe() faktiskt rknar ut dessa korrekt.

Mean: $\bar{x} = \frac{\sum x_i}{n}$

Varians: $\sigma^2 = \frac{\sum (x_i - \bar{x})}{n-1}$

Standardavvikelse = $ \sigma = \sqrt s^2 $

Count visar antalet vrden i kolumnen. Min & max vrden visar helt enkelt den minsta och den strsta siffran i kolumnen, och dr emellan visas percentilerna, 25 50 och 75. Dessa kan i viss mn ge ledtrdar till var de flesta vrdena ligger.

```python
total = sum(df['Sales'])
mean = total / df['Sales'].count()
deviations = [abs(row-mean)**2 for row in df['Sales']]
s2 = sum(deviations) / (df['Sales'].count() -1) #variance
s = np.sqrt(s2)

print(f"Mean: {mean}")
print(f"Standard deviation : {s}")
```

Hr ser vi ven att eftersom sigma^2(variansen) i mitt exempel rknas med bessels korrigering och standardavvikelsen blir densamma som nr vi rknar med pandas, s har vi pvisat att pandas anvnder sig av Bessels korrigering. Utan korrigeringen blir sigma = ungefr 20, vilket inte stmmer verens med pandas resultat.
