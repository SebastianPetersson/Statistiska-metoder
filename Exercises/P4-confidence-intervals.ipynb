### Exercise 1
The heat capacity for a liquid is measured as follows:

|kcal/gC°|.645|.654|.640|.627|.626|.649|.629|.631|.643|.633|.646|.630|.634|.631|.651|.659|.638|.645|.655|.624|.658|.658|.658|.647|.665|
|--------|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|

Past experience indicates that $\sigma=.01$.


**A**) Evaluate $\bar{X}$, thereby obtaining an unbiased point estimate for $\mu$.


```python
import numpy as np

measurements = [.645, .654, .640, .627, .626, .649, .629, .631, .643, .633,
                 .646, .630, .634, .631, .651, .659, .638, .645, .655, .624,
                   .658, .658, .658, .647, .665,]
x_mean = np.mean(measurements)
x_mean.round(5)
```

**B**) Assume that $X$ is normally distributed. Find a 95% confidence interval for $\mu$.

EGNA NOTES:

Stickprovsmedlet:  
  
$ \bar{X} = \frac{1}{n} \sum_{i=1}^{n}X_i$    
  
    
Standard Error:  
  
$ SE = \frac{\sigma}{\sqrt{n}}$  
  
Z-score:  
  
$\text{Z} = \frac{x - \mu}{\sigma}$

Margin:  
  
$\text{margin} = z_{\alpha/2} \cdot SE$

Konfidensintervall:  
  
$CI_{95\%} = \bar{X} \pm \text{margin}$

```python
import scipy.stats as st

n = len(measurements)
sigma = 0.01
alpha = 0.05 # --> 95% konfidensintervall
z = st.norm.ppf(1 - alpha/2)
se = sigma / np.sqrt(n)
margin = z * se

upper = x_mean + margin
lower = x_mean - margin

print(f"X_mean: {x_mean.round(5)}\nZ-score: {z.round(5)}\nStandard Error: {se}\nMargin: {margin.round(5)}\nUpper: {upper.round(5)}\nLower: {lower.round(5)} ")
```

**C**) Would you expect a 90% confidence interval for $\mu$ based on these data to be longer or shorter than the interval in b)?


*Jag förväntar mig att ett konfidensintervall på 90% kommer vara **mindre/kortare** än 95%, eftersom 90% gör spannet snävare.*

```python
alpha90 = 0.1
z_90 = st.norm.ppf(1 - alpha90/2)
z_90
```

**D**) Would you expect a 99% confidence interval for $\mu$ based on these data to be longer or shorter than the interval in b)? Verify by finding a 99% confindence interval on $\mu$.


*Jag förväntar mig att 99% konfidensintervall kommer vara **längre/bredare** eftersom den ökar chansen att datan ska ligga inom intervallen.*

```python
alpha99 = 0.01
z_99 = st.norm.ppf(1 - alpha99/2)
print(f"Z_90: {z_90.round(5)} < Z_95: {z.round(5)} < Z_99: {z_99.round(5)}")
```

### Exercise 3

Sulfur dioxide causes acid rain. These data were obtained on the concentration in a forest area thought to have been damaged by acid rain:
|µg/m³|52.7|43.9|41.7|71.5|47.6|55.1|62.2|56.5|33.4|61.8|54.3|50.0|45.3|63.4|53.9|65.5|66.6|70.0|52.4|38.6|46.1|44.4|60.7|56.4|
|-----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|






**A**) Compute $\bar{x}$, $s^2$ and $s$, the sample mean, sample variance and sample deviation.

```python
sulfur_data = [52.7, 43.9, 41.7, 71.5, 47.6, 55.1, 62.2, 56.5, 33.4, 61.8, 54.3, 50.0, 45.3, 63.4, 53.9, 65.5, 66.6, 70.0, 52.4, 38.6, 46.1, 44.4, 60.7, 56.4]

X_ = np.mean(sulfur_data)
s2 = np.var(sulfur_data, ddof=1)
s = np.sqrt(s2)
X_, s2, s
```

**B**) Compute a 95% confidence interval on the mean. 


```python
n = len(sulfur_data)
alpha = 0.05
t = st.t.ppf(1- alpha/2, df=n-1) #pop-standardavvikelse är okänd, därav T-värde istället för Z!
se = s / np.sqrt(n)
margin = t * se

upper = X_ + margin
lower = X_ - margin

lower, upper
```
