```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng(37)
integers = rng.integers(100, size=100)
uniform = rng.uniform(size=100)
normal = rng.normal(size=100)
binomial = rng.binomial(n=10,p=.3,size=100)
gamma = rng.gamma(shape=1.0, scale=0.9, size=100)
geometric= rng.geometric(p=0.31, size=100)
```

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(2, 3, figsize=(12,8))

#cumulative=True skapar en ogive, istllet fr att datan delas in i olika bins som vanligt.
sns.histplot(integers, ax=axes[0,0], stat="frequency", element="bars")
sns.histplot(uniform, ax=axes[0,1], stat="frequency", element="bars")
sns.histplot(normal, ax=axes[0,2], stat="frequency", element="bars")
sns.histplot(binomial, ax=axes[1,0], stat="frequency", element="bars")
sns.histplot(gamma, ax=axes[1,1], stat="frequency", element="bars")
sns.histplot(geometric, ax=axes[1,2], stat="frequency", element="bars")

axes[0,0].set_title("integers")
axes[0,1].set_title("uniform")
axes[0,2].set_title("normal")
axes[1,0].set_title("binomial")
axes[1,1].set_title("binomial")
axes[1,2].set_title("binomial")
plt.show()
```

```python
listed = [integers, uniform, normal, binomial, gamma, geometric]
titles = ['Integers', 'Uniform', 'Normal', 'Binomial', 'Gamma', 'Geometric']

fig1, axes1 = plt.subplots(2, 3, figsize=(12,8))
fig1.suptitle('Histograms', fontsize = 20)
axes1 = axes1.flatten()

for data, ax in zip(listed, axes1):
    sns.histplot(data, ax=ax, stat="frequency", element="bars")

for title, ax in zip(titles, axes1):
    ax.set_title(title)

#_______________________________________________________________________________________

fig2, axes2 = plt.subplots(2, 3, figsize=(12,8))
fig2.suptitle('Cumulative Histograms (Ogives) with ECDF Curves', fontsize = 20)
axes2 = axes2.flatten()

for data, ax in zip(listed, axes2):
    sns.histplot(data, ax=ax, cumulative=True, stat='density', element='step') #stat kan vara = frequency ocks, men d syns inte ecdf-linjen alls bra eftersom dess span r 0-1.

for data, ax in zip(listed, axes2):
    sns.ecdfplot(data, ax=ax)

for title, ax in zip(titles, axes2):
    ax.set_title(title)
```
